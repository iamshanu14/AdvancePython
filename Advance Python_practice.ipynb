# List Comprehension 

List comprehensions provide a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in
lists.

```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i**2)
    return lst1
    
```

```python
lst_square([1,2,3,4,5])
```

```python
l2=[1,3,2,4,65,6,7,8,9,90,100,34,67,44,33,34,523]
```

```python
[i**2 for i in l2]
```

```python
[i**2 for i in l2 if i%2==0]
```

```python
[i**2 for i in l2 if i%2!=0]
```

# Python OOPS Inheritance

```python
### All class variables are Public
### This is the blueprint of the Car
class Car():
    def __init__(self,window,door,enginetype):
        self.window=window
        self.door=door
        self.enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

```python
### All class variables are Private "_"
### This is the blueprint of the Car
class Car_private():
    def __init__(self,window,door,enginetype):
        self._window=window
        self._door=door
        self._enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

```python
### All class variables are Protected "__"
# This is the blueprint of the Car
class Car_protected():
    def __init__(self,window,door,enginetype):
        self.__window=window
        self.__door=door
        self.__enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

# Class Methods and Class Variables

```python
class Car:
    base_price = 1000000 # variable
    def __init__(self,door,window,power):
        self.door=door
        self.window=window
        self.power=power

    def what_base_price(self):
        print("the base price is: {}".format(self.base_price))
    @classmethod
    def revise_base_price(cls,inflation):
        cls.base_price=cls.base_price+cls.base_price*inflation
        
```

# Static Methods

```python
import datetime
now = datetime.datetime.now()
```

```python
class Car:
    base_price = 1000000 # variable
    def __init__(self,door,window,power):
        self.door=door
        self.window=window
        self.power=power

    def what_base_price(self):
        print("the base price is{}".format(self.base_price))
    @classmethod
    def revise_base_price(cls,inflation):
        cls.base_price=cls.base_price+cls.base_price*inflation
    @staticmethod
    def check_year():
        if now.year==2025:
            return True
        else:
            return False
```

```python
car2=Car(2,5,3000)
```

```python
car2.base_price
```

```python
car2.check_year()
```

# OOPS Multiple Inheritance in Python

```python
class A:
    def method1(self):
        print("A method is called")
```

```python
class B(A):
    def method1(self):
        print("B method is called")
```

```python
class C(A):
    def method1(self):
        print("C method is called")
```

```python
class D(B,C):
    def method1(self):
        print("D method is called")
        B.method1(self)
        C.method1(self)
        A.method1(self)
```

```python
d=D()
```

```python
d.method1()
```

```python
B.method1(d)
```

```python
d.method1()
```
