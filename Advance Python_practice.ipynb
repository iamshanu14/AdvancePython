# List Comprehension 

List comprehensions provide a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in
lists.

```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i**2)
    return lst1
    
```

```python
lst_square([1,2,3,4,5])
```

```python
l2=[1,3,2,4,65,6,7,8,9,90,100,34,67,44,33,34,523]
```

```python
[i**2 for i in l2]
```

```python
[i**2 for i in l2 if i%2==0]
```

```python
[i**2 for i in l2 if i%2!=0]
```

```python
try:
    #code block where exception can occur
    a= int(input("Enter the 1st Number:"))
    b= int(input("Enter the 2nd Number:"))
    c= a//b
    d=a*b
    e=a+b
    print("Division = ",c)
    print("Multiplication =",d)
    print("Additon =",e)
    
except NameError:
    print("the user has not defined the variabls")
except ZeroDivisionError:
    print("please provide number greater than 0")
except TypeError:
    print("Try to make datatype similar")
except Exception as ex:
    print(ex)
```

###    try else

```python
try:
    #code block where exception can occur
    a= int(input("Enter the 1st Number:"))
    b= int(input("Enter the 2nd Number:"))
    c= a//b
    d=a*b
    e=a+b
except NameError:
    print("the user has not defined the variabls")
except ZeroDivisionError:
    print("please provide number greater than 0")
except TypeError:
    print("Try to make datatype similar")
except Exception as ex:
    print(ex)
else:
    print("Division = ",c)
    print("Multiplication =",d)
    print("Additon =",e)
```

### try else finally

```python
try:
    #code block where exception can occur
    a= int(input("Enter the 1st Number:"))
    b= int(input("Enter the 2nd Number:"))
    c= a//b
except NameError:
    print("the user has not defined the variabls")
except ZeroDivisionError:
    print("please provide number greater than 0")
except TypeError:
    print("Try to make datatype similar")
except Exception as ex:
    print(ex)
else:
    print("Division = ",c)
finally:
    print("the execution is done")
```

# Python OOPS Inheritance

```python
### All class variables are Public
### This is the blueprint of the Car
class Car():
    def __init__(self,window,door,enginetype):
        self.window=window
        self.door=door
        self.enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

```python
### All class variables are Private "_"
### This is the blueprint of the Car
class Car_private():
    def __init__(self,window,door,enginetype):
        self._window=window
        self._door=door
        self._enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

```python
### All class variables are Protected "__"
# This is the blueprint of the Car
class Car_protected():
    def __init__(self,window,door,enginetype):
        self.__window=window
        self.__door=door
        self.__enginetype=enginetype
    def drive(self):
        print("The person Drive the car")
```

```python
car = Car(5,4,"petrol")
```

```python
car.door
```

```python
dir(car)
```

```python
car.drive()
```

```python
class audi(Car):
    def __init__(self,window,door,enginetype,enableai):
        super().__init__(window,door,enginetype)
        self.enableai=enableai
    def self_driving(self):
        print("Audi supports self driving")
```

```python
audiq7= audi(4,5,"electric",True)
```

```python
dir(audiq7)
```

```python
audiq7.door
```

```python
audiq7.self_driving()
```

```python
audiq7.drive()
```

# Python OOPS Magic methods in Classes

```python
### All class variables are Public
### This is the blueprint of the Car
class Car():
    def __new__(self,window,door,enginetype):
        print("The object is getting initialised")
    def __init__(self,window,door,enginetype):
        self.window=window
        self.door=door
        self.enginetype=enginetype
    def __str__(self):
        return "The object has been intialised"
    def __sizeof__(self):
        return "This display the size of the object"
    def drive(self):
        print("The person Drive the car")
```

```python
c = Car(4,5,"Diesel")
```

# Python Assert

Python provides the assert statement to check if a given logical expression is true or false. Program execution proceeds only if the expression is true and raises the AssertionError when it is false. The following code shows the usage of the assert statement.

```python
10>10
```

```python
10>=10
```

```python
num = 10
assert num>10
```

```python
num = 10
assert num>=10
```

```python
try:
    num=int(input("Enter a Number:"))
    assert num%2==0
    print("the number is even")
except AssertionError:
    print("Please Enter a Even Number")
```
